```python
def toint(a):
    if ' ' in a:
        s=a.split(' ')
        if len(s)==2:
            d=int(s[0])*1000+int(s[1])
            return d
        if len(s)==3:
            d=int(s[0])*1000000+int(s[1])*1000+int(s[2])
            return d
        
    elif '.' in a:
        s=a.split('.')
        if len(s)==2:
            d=int(s[0])*1000+int(s[1])
            return d
        if len(s)==3:
            d=int(s[0])*1000000+int(s[1])*1000+int(s[2])
            return d
    else:
        return int(a)
toint('12 200')
toint('121.290')
toint('9.090.990')
```
